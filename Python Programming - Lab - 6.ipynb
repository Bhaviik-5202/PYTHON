### 01) WAP to find sum of all the elements in a List.

```python
n = int(input("Enter number OF Elements: "))
lt = []
for i in range(n):
    x = int(input("Enter Element: "))
    lt.append(x)

sum = 0
for i in lt:
    sum += i
print("List =", lt)
print("Sum OF All Element :", sum)
```

## 02) WAP to find largest element in a List.

```python
lt = [12, 45, 7, 89, 30]
largest = lt[0]
for i in lt:
    if i > largest:
        largest = i

print("List : ", lt)
print("Largest Element in List :", largest)
```

##### 08) WAP to print even numbers in a list.

```python
lt = [10, 15, 20, 25, 30, 35, 40, 45, 50]
res = []
for i in lt:
    if i % 2 == 0:
        res.append(i)
print("Original List : ", lt)
print("Even Number List :", res)
# OR even = [i for i in lst if i % 2 == 0]
```

### 09) WAP to count unique items in a list.

```python
lst = [10, 20, 10, 30, 20, 40, 50, 40]
unique = []
for i in lst:
    if i not in unique:
        unique.append(i)

print("List =", lst)
print("Unique items =", unique)
print("Count of unique items =", len(unique))
```

### 11) WAP to print all odd numbers in a given range.

```python
pos1 = int(input("Enter Starting Range :"))
pos2 = int(input("Enter Ending Range :"))
oddLt = []
for i in range(pos1, pos2):
    if i % 2 != 0:
        oddLt.append(i)
print("Odd Number List :", oddLt)
```

### 12) WAP to count occurrences of an element in a list.

```python
lt = [10, 10, 20, 20, 30, 30, 30, 40, 50]
key = int(input("Enter Element :"))
count = 0
for i in lt:
    if key == i:
        count += 1
print("Original List : ", lt)
print("Element =", key)
print("Occurrences =", count)
```

### 14) WAP to extract elements with frequency greater than K.

```python
lt = [10, 10, 10, 20, 30, 40, 50, 60]
k = int(input("Enter Frequency :"))
res = []
for i in lt:
    if lt.count(i) > k and i not in res:
        res.append(i)

print("origina List :", lt)
print("K =", k)
print("Elements with frequency greater than K =", res)
```

### 16) WAP to create a new list (fruit whose name starts with 'b') from the list of fruits given by user.

```python
# lt = ['banana', 'apple', 'watermelon', 'orange']
n = int(input("Enter number of fruits: "))
fruits = []
for i in range(n):
    fruits.append(input("Enter fruit name: "))

b_fruits = [f for f in fruits if f.startswith('b')]

print("Fruits starting with 'b' =", b_fruits)
```
