### 12) WAP to find whether the given number is Prime or not.

```python
n = int(input("Enter Number:"))
isPrime = 1

if n <= 1:
    print(f"{n} is Not Prime")

for i in range(2, n):
    if(n % i == 0):
        isPrime = 0
        break

if(isPrime):
    print(f"{n} is Prime Number.")
else:
    print(f"{n} is Not Prime Number.")
```

### 13) WAP to print sum of digits of given number.

```python
n = int(input("Enter Number :"))
sum = 0
for i in range(1,n+1):
    rem = n % 10
    sum += rem
    n //= 10

print(f"Sum OF Digit : {sum}")
```

### 14) WAP to check whether the given number is Palindrome or not.

```python
n = int(input("Enter Number:"))
isPal = 0
temp = n
while temp != 0:
    isPal = isPal * 10 + temp % 10
    temp //= 10
if n ==  isPal:
    print(f"{n} is Palindrome Number.")
else:
    print(f"{n} is Not Palindrome Number.")
```

### 15) WAP to check whether the given number is an Armstrong Number or not.

```python
n = input("Enter Number:")
isArm = 0
digit = len(n)
n = int(n)
temp = n
while temp != 0:
    rem = temp % 10
    isArm += rem ** digit
    temp //= 10
if n == isArm:
    print(f"{n} is Armstrong Number.")
else:
    print(f"{n} is Not Armstrong Number.")
```

### 16) WAP to print all the perfect numbers between 1 to n.

```python
n1 = int(input("Enter Number: "))
n2 = int(input("Enter Number: "))
print(f"{n1} To {n2} Perfect Number : ")
for i in range(n1,n2+1):
    sum = 0
    for j in range(1,i):
        if i % j == 0:
            sum += j
    if sum == i:
        print(i)
```
